```python
#generators
#they calculate and print the elements of the list, and will not save them in the memory
def prine(n):
    for i in range (2,n):
        for j in range (2,i):
            if i%j==0:
                break
            else:
                yield i
p=prine(50)
print(next(p))
print(next(p))
print(list(p))
            
```

```python
#iterables

class first100:
    def __init__(self):
        self.number=0
    def __next__(self):
        if self.number<100:
            current=self.number
            self.number+=1
            return current
        else:
                raise StopIteration()
#generator tho produce aina list ni,all at once access cheydanki ,we use Iterables 
#without storing
class iterable:
    def __iter__(self):
          return first100()

    
for i in iterable():
      print(i)
```

```python
#iterables
#from  the same class
class first100:
    def __init__(self):
        self.number=0
    def __next__(self):
        if self.number<100:
            current=self.number
            self.number+=1
            return current
        else:
                raise StopIteration() 
    def __iter__(self):
          return self

    
for i in first100():
      print(i)
```
